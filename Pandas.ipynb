```
Pandas is an open-source Python library used for data manipulation and data analysis, especially with structured data in the form of tables (rows and columns).
PANDAS = Panel Data Analysis.
“Panel Data” refers to data that has multiple dimensions (like time, rows, columns).
Pandas was designed to handle this type of structured data efficiently.


Pandas is a library used to:

- Load data
- Clean data
- Analyze data
- Manipulate data

It works especially well with tables (like Excel sheets or CSV files).

### <mark>**Series : A series is a one dimensional labelled vertical array**<mark/>

Series is a single column of data with label(index)

````python
import numpy as np
import pandas as pd
````

````python
labels = ['a','b','c']
my_list = [10,20,30]
arr = np.array([10,20,30])
d = {1:10,2:20,3:30}
````

````python
s1 = pd.Series(my_list)
s1
````

````python
s2 = pd.Series(my_list,index=labels)
s2
````

````python
s3 = pd.Series(arr)
s3
````

````python
s4 = pd.Series(d)
s4
````

#### <mark>**DataFrame : A DataFrame is a two-dimensional (2D) data structure in Pandas that stores data in rows and columns (like a table).**<mark/>

DataFrame is collection of multiple series(column) forming a table.

````python
data = {
    "Name": ["Rohan", "Amit", "Neha"],
    "Age": [21, 22, 20],
    "City": ["Delhi", "Mumbai", "Pune"]
}
df = pd.DataFrame(data)
print(df)
````

````python
data = [
    ["Rohan", 21, "Delhi"],
    ["Amit", 22, "Mumbai"],
    ["Neha", 20, "Pune"]
]
columns_name = ["Name","Age","city"]
df2 = pd.DataFrame(data,columns=columns_name)
df2
````

### <mark>**Selection and indexing of Columns**<mark/>

````python
data = {
    "Name": ["Rohan", "Amit", "Neha"],
    "Age": [21, 22, 20],
    "City": ["Delhi", "Mumbai", "Pune"]
}
df3 = pd.DataFrame(data)
df3
````

````python
# select single ccolumns
df3["Name"]
````

````python
# selecting multiple ccolumns_name
df3[["Name","Age"]]
````

#### <mark>**Selecting column using dot**<mark/>

````python
df3.Name
````

````python
df3.City
````

**<mark>Note: Works only if column name has no space<mark/>**

### **<mark>Adding a new column<mark/>**

````python
df3
````

````python
df3["Designation"] = ["data scientist","software engineer","ML engineer"]
df3
````

````python
df3["Experience"] = ["Fresher","1yr","3yrs"]
df3
````

### <mark>**Removing a column**<mark/>

````python
df3.drop("Designation",axis=1) # you have to give axis=1 to remove column because defaut value of axis is 0 it will try to remove row name with designation
````

````python
df3 # you can see it does not affect the original dataframe it to delete the column for original dataframe we have to use inplace attribute
````

````python
df3.drop("Designation",axis=1,inplace=True)
````

````python
df3 # now you can see Designation is removed permanently from dataframe
````

### **Removing multiple columns**

````python
df3.drop(["Name","Age"],axis=1)
````
