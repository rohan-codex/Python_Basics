```
Pandas is an open-source Python library used for data manipulation and data analysis, especially with structured data in the form of tables (rows and columns).
PANDAS = Panel Data Analysis.
“Panel Data” refers to data that has multiple dimensions (like time, rows, columns).
Pandas was designed to handle this type of structured data efficiently.

- Load data
- Clean data
- Analyze data
- Manipulate data

### <mark>**Series : A series is a one dimensional labelled vertical array**<mark/>

Series is a single column of data with label(index)

````python
import numpy as np
import pandas as pd
````

````python
labels = ['a','b','c']
my_list = [10,20,30]
arr = np.array([10,20,30])
d = {1:10,2:20,3:30}
````

````python
s1 = pd.Series(my_list)
s1
````

````python
s2 = pd.Series(my_list,index=labels)
s2
````

````python
s4 = pd.Series(d)
s4
````

#### <mark>**DataFrame : A DataFrame is a two-dimensional (2D) data structure in Pandas that stores data in rows and columns (like a table).**<mark/>

DataFrame is collection of multiple series(column) forming a table.

````python
data = {
    "Name": ["Rohan", "Amit", "Neha"],
    "Age": [21, 22, 20],
    "City": ["Delhi", "Mumbai", "Pune"]
}
df = pd.DataFrame(data)
print(df)
````

````python
data = [
    ["Rohan", 21, "Delhi"],
    ["Amit", 22, "Mumbai"],
    ["Neha", 20, "Pune"]
]
columns_name = ["Name","Age","city"]
df2 = pd.DataFrame(data,columns=columns_name)
df2
````

### <mark>**Removing a column**<mark/>

````python
df3.drop("Designation",axis=1) # you have to give axis=1 to remove column because defaut value of axis is 0 it will try to remove row name with designation
````

````python
df3 # you can see it does not affect the original dataframe it to delete the column for original dataframe we have to use inplace attribute
````

````python
df3.drop("Designation",axis=1,inplace=True)
````

````python
df3 # now you can see Designation is removed permanently from dataframe
````

### **Removing multiple columns**

````python
df3.drop(["Name","Age"],axis=1)
````

#### <mark>**Selecting Rows**<mark/>

<mark>**loc : loc is a label-based indexing method in Pandas used to select rows and columns using their names (labels).**<mark/>

**loc[row, column]**

````python
import numpy as np
import pandas as pd
data = {
    "Name": ["Rohan", "Amit", "Neha"],
    "Age": [21, 22, 20],
    "City": ["Delhi", "Mumbai", "Pune"]
}
df4 = pd.DataFrame(data)
df4
````

````python
df4.loc[0]
````

````python
df4.loc[0,"Name"]
````

````python
df4.loc[: ,"Name"] # it selects all rows(:)
````

````python
df4.loc[[0,1]] # selecting multiple rows
````

<mark>**iloc : iloc is a position-based indexing method in Pandas used to select rows and columns using their integer index positions.**<mark/>

#### <mark>**Selecting subsets of row and column**<mark/>

````python
df4
````

````python
df4.loc[[0,1]][["Name","Age"]]
````

<mark>**Conditional Selection**<mark/>

````python
df4
````

````python
df4[df4["Age"]>21]
````

````python
df4[(df4["Age"]>20)&(df4["City"]=="Delhi")]
````

#### <mark>**Missing Data**<mark/>

```
Missing data means some values are not present in the dataset.
In Pandas, missing values are shown as:
NaN (Not a Number).

````python
import pandas as pd
import numpy as np
data = {
    "Name": ["Rohan", "Amit", "Neha"],
    "Age": [21, np.nan, 20],
    "City": ["Delhi", "Mumbai", np.nan ]
}
df = pd.DataFrame(data)
df
````

#### **<mark>Checking Missing Data**<mark/>

**<mark>df.isna()/df.isnull() : Returns a boolean DataFrame showing True for missing values (NaN) and False otherwise.<mark/>**

````python
df.isna()
````

````python
df.isnull()
````

**<mark>isna().sum() : Counts total missing values in each column (or row if axis is changed)**<mark/>
```
Works on:
Columns (default)
Rows → axis=1

````python
df.isna().sum() 
````

````python
df.isna().sum(axis=1)
````

<mark>**isna().any() :  checks if there is at least one missing value (NaN) .Returns True if any missing value is present, otherwise False.**<mark/>
```
Works on:
Columns (default)
Rows → axis=1

````python
df.isna().any()
````

````python
df.isna().any(axis=1)
````

**<mark>notna() / notnull() : Returns True for non-missing values and False for missing values.<mark/>**

````python
df.notna()
````
