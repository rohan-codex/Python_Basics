# Dictionary :
A dictionary is a data structure that stores data in key-value pairs.It is
- mutable(can be changed)
- unordered(before Python 3.7)
- Indexed by keys(Not by position) 

## Syntax 
```python
dictionary_name = {
    key1: value1,
    key2: value2
}
```

# Example

````python
student = {
    "name": "Rohan",
    "age": 21,
    "marks": 89
}
````

## Characteristics of dictionary
1.<mark>Keys must be unique</mark>
   

````python
d = {"a": 1, "a": 2}
print(d)
````

**Duplicate keys overwrite previous value.**

2. <mark>Keys must be immutable</mark>

Allowed:
- string
- number
- tuple

Not allowed:
- list
- dictionary

3.<mark>Values can be anything:</mark>

- string
- number
- list
- tuple
- dictionary
even another dictionary

exmple

````python
data = {
    "name": "Virat",
    "scores": [100, 85, 76],
    "stats": {"matches": 250, "runs": 12000}
}
````

3️. <mark>Accessing Values</mark>

Method 1: Using Key:
```python
student["name"]
```
Method 2: using .get()
```python
student.get("name")
```
Difference:
- [] gives error if key not found
- .get() returns None if key not found

4. <mark>Adding and Updating values</mark>

   adding new key:
   ```pthon
   student["city"] = "Delhi"
   ```
   updating value:
   ```python
   student["marks"] = 95
   ```
   
5. <mark>Deleting Elements</mark>
   ```python
   del student["age"]
   ```
OR
   ```pyhton
   student.pop("age")
   ```

6. <mark>Looping through Dictionary</mark>

   Loop through Keys:
   ```python
   for key in student:
    print(key)
   ```
   
   Loop through values:
   ```python
   for value in student.values():
    print(value)
   ```
   
   Loop through key and value:
   ```python
   for key, value in student.items():
    print(key, value)
   ```

8. <mark>Important Dictionary Methods</mark>

- .keys() - Returns all keys
- .values() - Returns all values
- .items() - Returns key-value pairs
- .update()	- Update dictionary
- .pop() -	Remove specific key
- .clear() - Remove all elements

8. <mark>Nested dictionary</mark>

 it is a dictionary inside a dictionary
example:
```python
ipl = {
    "RCB": {
        "captain": "Faf",
        "titles": 0
    },
    "MI": {
        "captain": "Rohit",
        "titles": 5
    }
}
```
Accessing nestes values:
```python
ipl["MI"]["titles"]
```

- Iterable -	Example
- List -	[1, 2, 3]
- Tuple - 	(1, 2, 3)
- String -	"Python"
- Dictionary -	{"a": 1, "b": 2}
- Set -	{1, 2, 3}
- Range	- range(5)

## 1. sum() Function :
sum() is used to add all elements in an iterable (like list, tuple, set).

### Syntax :
```python
sum(iterable)
```


#### Example

````python
student = [1,2,3,4,5,6]
total = sum(student)
print(total)
````

````python
student = {
    "maths" : 67,
    "science" : 65,
    "english" : 45
}
total = sum(student.values())
print(total)
````

````python
#Given employee salary data, find: Total salary , Average salary ,Employee with highest salary
employees = [
    {"name": "Aman", "salary": 50000},
    {"name": "Riya", "salary": 60000},
    {"name": "John", "salary": 55000}
]
total_salary = 0
highest_salary = 0
top_employee = ""

for emp in employees:
    total_salary += emp["salary"]

    if emp["salary"] > highest_salary:
        highest_salary = emp["salary"]
        top_employee = emp["name"]

average_salary = total_salary / len(employees)

print("Total Salary:", total_salary)
print("Average Salary:", average_salary)
print("Top Employee:", top_employee)
````

````python
# frequency counter
data = [1,1,2,2,2,2,3,3,3,4,4,5,5]
frequency={}
for item in data:
    if item in frequency:
        frequency[item]=frequency[item]+1
    else:
        frequency[item]=1
print(frequency)
````

````python
words = ["data", "science", "data", "python", "data", "python"]
marks = [90, 85, 90, 70, 85, 90, 70]

def frequency_counter(data):
    freq = {}

    for item in data:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1

    return freq

print("Word Frequency:", frequency_counter(words))
print("Marks Frequency:", frequency_counter(marks))
````

````python
marks = {
    "Rohan": 85,
    "Amit": 92,
    "Neha": 78,
    "Priya": 92,
    "Rahul": 85
}

# 1. Print all student names
print("Students:")
for student in marks.keys():
    print(student)

# 2. Print all marks
print("\nMarks:")
for score in marks.values():
    print(score)

# 3. Find the highest mark
highest_mark = max(marks.values())
print("\nHighest Mark:", highest_mark)

# 4. Count students scoring more than 80
count = 0
for score in marks.values():
    if score > 80:
        count += 1

print("Students scoring more than 80:", count)
````

````python
sales = ["apple", "banana", "apple", "orange", "banana", "apple"]

freq = {}

# 1. Create frequency dictionary
for item in sales:
    if item in freq:
        freq[item] += 1
    else:
        freq[item] = 1

# 2. Print frequency dictionary
print("Product Frequency:", freq)

# 3. Product sold maximum times
most_sold = max(freq, key=freq.get)
print("Most Sold Product:", most_sold)
````

````python
ages = [18, 22, 25, 17, 30, 22, 18, 25, 30, 30]

age_groups = {
    "teen": 0,
    "young": 0,
    "adult": 0
}

for age in ages:
    if age < 20:
        age_groups["teen"] += 1
    elif age < 30:
        age_groups["young"] += 1
    else:
        age_groups["adult"] += 1

print(age_groups)
````
