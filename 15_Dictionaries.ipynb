# Dictionary :
A dictionary is a data structure that stores data in key-value pairs.It is
- mutable(can be changed)
- unordered(before Python 3.7)
- Indexed by keys(Not by position) 

## Syntax 
```python
dictionary_name = {
    key1: value1,
    key2: value2
}
```

# Example

````python
student = {
    "name": "Rohan",
    "age": 21,
    "marks": 89
}
````

## Characteristics of dictionary
1.<mark>Keys must be unique</mark>
   

````python
d = {"a": 1, "a": 2}
print(d)
````

**Duplicate keys overwrite previous value.**

2. <mark>Keys must be immutable</mark>

Allowed:
- string
- number
- tuple

Not allowed:
- list
- dictionary

3.<mark>Values can be anything:</mark>

- string
- number
- list
- tuple
- dictionary
even another dictionary

exmple

````python
data = {
    "name": "Virat",
    "scores": [100, 85, 76],
    "stats": {"matches": 250, "runs": 12000}
}
````

3️. <mark>Accessing Values</mark>

Method 1: Using Key:
```python
student["name"]
```
Method 2: using .get()
```python
student.get("name")
```
Difference:
- [] gives error if key not found
- .get() returns None if key not found

4. <mark>Adding and Updating values</mark>

   adding new key:
   ```pthon
   student["city"] = "Delhi"
   ```
   updating value:
   ```python
   student["marks"] = 95
   ```
   
5. <mark>Deleting Elements</mark>
   ```python
   del student["age"]
   ```
OR
   ```pyhton
   student.pop("age")
   ```

6. <mark>Looping through Dictionary</mark>

   Loop through Keys:
   ```python
   for key in student:
    print(key)
   ```
   
   Loop through values:
   ```python
   for value in student.values():
    print(value)
   ```
   
   Loop through key and value:
   ```python
   for key, value in student.items():
    print(key, value)
   ```

8. <mark>Important Dictionary Methods</mark>

- .keys() - Returns all keys
- .values() - Returns all values
- .items() - Returns key-value pairs
- .update()	- Update dictionary
- .pop() -	Remove specific key
- .clear() - Remove all elements

8. <mark>Nested dictionary</mark>

 it is a dictionary inside a dictionary
example:
```python
ipl = {
    "RCB": {
        "captain": "Faf",
        "titles": 0
    },
    "MI": {
        "captain": "Rohit",
        "titles": 5
    }
}
```
Accessing nestes values:
```python
ipl["MI"]["titles"]
```

9. <mark>Importance of Dictionary in data science</mark>

(i). **JSON data - APIs return data in dictionary format.**

example:
```pyhton
{
  "player": "Virat",
  "runs": 5432
}
```

(ii). **Dataset Row Representation - Each row in dataset can be represented as:**

example:
```python
row = {
    "match_id": 101,
    "team": "RCB",
    "runs": 180
}
```

(iii). **Counting Frequency** 



````python
names = ["A", "B", "A", "C", "B", "A"]

freq = {}

for name in names:
    if name in freq:
        freq[name] += 1
    else:
        freq[name] = 1

print(freq)
````

(iv). **Used Internally in Pandas - A DataFrame is like:**
```python
data = {
    "Name": ["A", "B", "C"],
    "Marks": [90, 85, 88]
}
```

## NOTE :
**A dictionary is a mutable, unordered data structure that stores data in key–value pairs. Keys must be unique and immutable. Dictionaries are widely used in data science for handling JSON data, dataset rows, frequency counting, and structured data storage.**
