# Python Functions

## What is a function?

A function is a block of reusable code that performs a specific task.


## Syntax of a Function

```python
def function_name(parameters):
    statement(s)
    return value


## Function without Parameters


````python
def greet():
    print("Hello, welcome to Python functions")

greet()
````

- This function does not take any input
- It only performs an action
- We call the function using its name followed by ()


## Function with Parameters


````python
def greet(name):
    print("Hello", name)

greet("Rohan")
````

````python
def add(a, b):
    print(a + b)

add(10, 20)
````

````python
def average(a, b, c):
    avg = (a + b + c) / 3
    print(avg)

average(4, 6, 8)
````

- Parameters are variables written in function definition
- Values passed during function call are arguments

## Function with Return


````python
def add(a, b):
    return a + b

result = add(5, 7)
print(result)
````

- `return` sends the result back to the function call
- The returned value can be stored in a variable
- `print` is used to display the returned value


### Without return

````python
def add(a, b):
    print(a + b)
````

You can only see the result

### With return

````python
def add(a, b):
    return a + b
````

You can store, reuse, calculate further

## Difference between print and return


### Example 1: Using print

````python
def multiply(a, b):
    print(a * b)

result = multiply(3, 4)
print(result)
````

- `print` only displays value
- Function returns `None` by default

### Example 2: Using return

````python
def multiply(a, b):
    return a * b

result = multiply(3, 4)
print(result)
````

- `print` is used only to display output
- `return` sends value back from function
- Functions without `return` return `None`


### Note :

- Use `print` to see the output
- Use `return` to use the output

### Real life example

Imagine you ask a friend:

“What is 10 + 20?”

`Case 1️`: Friend says the answer (print) He says: “30” You hear it, but you can’t use it for anything else.

`Case 2️`: Friend writes it on paper and gives it to you (return) He gives you a paper with 30 written. Now you can: Save it Use it again Add more things to it

If you want to USE the result later , use return If you just want to SHOW the result , use print

## Practice Problems

### Problem 1 : Write a function to check a number is even or odd


### Solution :

````python
def check(n):
    if n%2==0:
        print("even")
    else:
        print("odd")
check(3)
check(4)
    
````

### Problem 2 : Write a function to find sum of numbers from 1 to n


### Solution :


````python
def sum(n):
    total=0
    for i in range(1,n+1):
        total+=i
    print(total)
sum(2)
sum(3)
````

### Problem 3 : Write a function to find square of a number and return it

### Solution :

````python
def square(n):
    return n*n
print(square(6))
print(square(3))
````
