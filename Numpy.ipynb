### **<mark>NUMPY INDEXING AND SLICING OF VECTORS<mark/>**

In NumPy, vectors are just 1D arrays, and indexing/slicing is how you access or extract elements from them.

````python
import numpy as np
````

````python
arr = np.arange(1,31)
arr
````

#### **<mark>Positive Indexing<mark/>**

````python
print(arr[0])
print(arr[2])
print(arr[8])
````

#### <mark>**Negative Indexing<mark/>**

````python
print(arr[-1])
print(arr[-2])
print(arr[-15])
````

#### <mark>**Slicing (Access Multiple Elements)**<mark/>

**Syntax:**

```arr[start : end : step]```

````python
print(arr[0:8]) #it will go index 0 to 7
````

````python
print(arr[0:8:2]) #it will print every second value
````

````python
print(arr[:7]) # default value of start is 0
````

````python
print(arr[5:]) # default value of end is the last index + 1
````

````python
print(arr[:]) # it will print the vector
````

````python
print(arr[::2]) # step = 2 (skip one element)
````

````python
print(arr[::-1]) # it will reverse the vector
````

#### <mark> **View vs copy**<mark/>

**Slicing creates a view not a copy , changes affect the original array.**

````python
arr1 = np.arange(0,6)
print(arr1)

````

````python
slice_arr = arr1[0:3]
print(slice_arr)
````

````python
slice_arr[1]=10
print(arr1)
````

**to avoid this problem use copy**

````python
slice_arr = arr1[0:3].copy()
slice_arr[1] = 100
print(slice_arr)
print(arr1)
````

#### <mark>**Fancy Indexing : Access multiple specific indices**<mark/>

````python
arr2 = np.arange(10,31)
print(arr2)
````

````python
print(arr2[[0,3,5]])
````

### <mark>**NUMPY INDEXING AND SLICING OF MATRIX**<mark/>

In NumPy, a matrix is a 2D array, so indexing and slicing involve rows and columns.


````python
import numpy as np
arr3 = np.arange(1,31).reshape(6,5)
print(arr3)
````

#### <mark>**INDEXING IN MATRIX**<maark/>

**Syntax:**

``` arr[row,column]```

````python
print(arr3[0,0])
print(arr3[3,4])
````

**Negative Indexing**

````python
print(arr3[-1,-1])
print(arr3[-2,-5])
````

#### **<mark>SELECTING ENTRE ROW / COLUMN<mark/>**

````python
print(arr3[3]) # it will give row of index 3
````

OR

````python
print(arr3[3, :])
````

````python
print(arr3[:,3]) # it will give column of index 3
````

#### **<mark>SLICING ROWS AND COLUMNS<mark/>**

**Syntax:**


```arr[row_start:row_end , col_start:col_end]```

````python
print(arr3[0:2,1:3]) # it will start from row of index 0 and moves to row of index 2-1=1 same for column
````

````python
print(arr3[3:,3: ]) # it will start for row of index 0 and moves until last because we dont provvide end value same as for column
````

````python
print(arr3[:2,:]) # it will give first two rows because for selecting two rows you have select all columns so you dont give any index for column default values will do its job
````

````python
print(arr3[:,:2]) # it will give first two column same logic as before
````

#### <mark>**STEP SLICING<mark/>**

````python
arr3
````

````python
print(arr3[::2,::2])
````

Rows → ::2
- Take row 0
- Skip row 1
- Take row 2

Columns → ::2
- Take col 0
- Skip col 1
- Take col 2

````python
print(arr3[1::2,1:3])
````

#### <mark>**BOOLEAN INDEXING**<mark/>

**In NumPy, boolean indexing is a powerful way to filter data using conditions . It means selecting elements using True / False conditions.**

````python
arr3
````

````python
print(arr3>15)
````

````python
print(arr3[arr3>15])
````

**<mark>NOTE:<mark/>**

**First,it checks the condition on a matrix and gives a boolean maatrix as output then feed this as index in the original matrix name to fetch the values
and output will be 1D array**

````python
print(arr3[arr3<10]) # it will give the values less than 10 from the matrix arrr3
````

````python
print(arr3[arr3%2==0])
````

````python
print(arr3[(arr3>5) & (arr3<15)])
````

#### <mark>**Modify values using Boolean Indexing**<mark/>

````python
arr3[arr3>25]=0 # it will modify those values which is greater than 25 and change to 0 . It is very useful in Data cleaning
print(arr3) 
````

````python
arr3[(arr3>5) & (arr3<10)] = 1
print(arr3)
````

### <mark>**BROADCASTING**<mark/>

**NumPy automatically makes arrays the same shape to perform operations . So you don’t need loops.**

````python
arr = np.array([1, 2, 3])
print(arr + 5)
````

NumPy internally does: 
``` [1 2 3] ```
``` [5 5 5] ```  ← expanded automatically

````python
a = np.array([1, 2, 3])
b = np.array([10])
print(a + b)
````

b is stretched to match a

````python
arr = np.array([
    [1, 2],
    [3, 4]
])

print(arr + 10)
````

10 added to every element

````python
arr = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

b = np.array([10, 20, 30])

print(arr + b)
````

What happened:
Matrix:
[[1 2 3]
 [4 5 6]]

Vector:
[10 20 30]

Becomes:
[[10 20 30]
 [10 20 30]]
