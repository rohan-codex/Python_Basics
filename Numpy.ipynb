## <mark>Numpy</mark> :
 Numpy stands for numerical python. It is a python library used for fast numerical and mathematical operations on data.
 Numpy helps us work with large amounts of numbers efficiently using arrays.

### <mark>Why Numpy is used :</mark>
Normal python lists are slow for numerical computation. Numpy provides arrays that are :
- faster
- memory efficient
- designed for scientific computing and data science.

### <mark>Core concept : Numpy array<mark/>
A numpy array is similar to a python list but much more powerful.

example:

```python
import numpy as np
arr = np.array([10,20,30,40])
print(arr)
```

- np → short name (alias) for NumPy,we can name whatever we want
- array() → a function that creates a NumPy array
- np.array() is a function used to convert a list or sequence into a NumPy array.

### <mark>Difference between List and NumPy array</mark>
Python List :

```python
num = [1,2,3]
```

NumPy array :


```python
num = np.array([1,2,3])
```

**but numpy array allows vector operations :**

```python
arr = np.array([1,2,3])
print(arr*2)
```

**this is very powerful for data analysis**
