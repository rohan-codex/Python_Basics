# Python Functions

## What is a function?

A function is a block of reusable code that performs a specific task.


## Syntax of a Function

```python
def function_name(parameters):
    statement(s)
    return value


## Function without Parameters


````python
def greet():
    print("Hello, welcome to Python functions")

greet()
````

- This function does not take any input
- It only performs an action
- We call the function using its name followed by ()


## Function with Parameters


````python
def greet(name):
    print("Hello", name)

greet("Rohan")
````

````python
def add(a, b):
    print(a + b)

add(10, 20)
````

````python
def average(a, b, c):
    avg = (a + b + c) / 3
    print(avg)

average(4, 6, 8)
````

- Parameters are variables written in function definition
- Values passed during function call are arguments

## Function with Return


````python
def add(a, b):
    return a + b

result = add(5, 7)
print(result)
````

- `return` sends the result back to the function call
- The returned value can be stored in a variable
- `print` is used to display the returned value


### Without return

````python
def add(a, b):
    print(a + b)
````

You can only see the result

### With return

````python
def add(a, b):
    return a + b
````

You can store, reuse, calculate further
